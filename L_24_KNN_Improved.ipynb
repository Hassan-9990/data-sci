# L_24: K-Nearest Neighbors (KNN)

KNN is a simple yet powerful classification algorithm based on feature similarity.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
# Load dataset
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
# Evaluate model
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred))
ConfusionMatrixDisplay.from_estimator(knn, X_test, y_test)
plt.title("KNN Confusion Matrix")
plt.grid(False)
plt.show()
```

## ℹ️ Key Points:
- No training phase (lazy learning)
- Accuracy sensitive to the value of k
- Works well on small datasets with well-separated classes

✅ Use KNN when interpretability and simplicity are important.
