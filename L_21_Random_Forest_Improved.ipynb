# L_21: Random Forest

Random Forest is an ensemble method that builds multiple decision trees and combines their outputs.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
# Load dataset
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
# Evaluate model
y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred))
ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)
plt.title("Confusion Matrix - Random Forest")
plt.grid(False)
plt.show()
```

```python
# Feature importance
import pandas as pd
import numpy as np
features = load_breast_cancer().feature_names
importances = rf.feature_importances_
sorted_idx = np.argsort(importances)[-10:]
plt.figure(figsize=(10,5))
plt.barh(features[sorted_idx], importances[sorted_idx])
plt.xlabel("Importance")
plt.title("Top 10 Feature Importances")
plt.grid(True)
plt.show()
```

## âœ… Key Takeaways
- Combines results of multiple decision trees to reduce overfitting.
- Provides feature importance scores.
- Excellent for both classification and regression.
